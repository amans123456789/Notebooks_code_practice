### DP

#### Recurrsion + Memo

```python
coins = [1,2,5,12,15]
amount = 11
n = len(coins)
```

```python
coins = [2]
amount = 3
n = len(coins)
```

```python
n = len(coins)
dp =[ [-1 for i in range(amount + 1)] for j in range(n+1)]
```

```python

def coin_change_dp(coins, amount,n):
    print(coins, n-1)

    if n == 0 or amount == 0:
        return 0
    
    if dp[n][amount] != -1:
        return dp[n][amount]
    
    if coins[n-1] > amount:
        dp[n][amount] = coin_change_dp(coins, amount,n-1)
        return dp[n][amount]
    
    dp[n][amount] = coin_change_dp(coins, amount - coins[n-1],n) + 1
    
    return dp[n][amount]
```

```python
coin_change_dp(coins, amount,n)
```

### Tabulation

```python
coins = [2]
amount = 3
n = len(coins)
```

```python
coins = [1,2,5]
amount = 11
n = len(coins)
```

```python
def coin_change_tab(coins, amount,n):
    
    dp = [[0 for x in range(amount + 1)] for x in range(n + 1)]
    for i in range(1,len(dp[0])):
        dp[0][i]=1000
    for i in range(1,n+1):
        for j in range(1,amount+1):
#             if n == 0 or amount == 0:
#                 dp[i][j] = 0
            if coins[i-1] <= j:
                dp[i][j] = min(dp[i-1][j],1 + dp[i][j-coins[i-1]] )
            else:
                dp[i][j] = dp[i-1][j]
                
    if(dp[len(coins)][amount]==1000):
        return -1
    
    return dp[n][amount]
```

```python
coin_change_tab(coins, amount,n)
```
