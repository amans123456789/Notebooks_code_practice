#### Recurrsion

```python
def knapsack(N, W, profit, weight):
    
    if N == 0:
        return 0
    if W < weight[0]:
        return knapsack(N - 1, W, profit[1:], weight[1:])
    
    profit_include_wt = knapsack(N - 1, W -weight[0] , profit[1:], weight[1:]) + profit[0]
    profit_exclude_wt = knapsack(N-1, W , profit[1:], weight[1:])

    return max(profit_include_wt,profit_exclude_wt )
```

```python
knapsack(N, W, profit, weight)
```

### DP

#### Recurrsion + Memo

```python
memo = {}
for i in range(N):
    memo[i] = 0
```

```python
memo = [[-1 for i in range(W+1)] for j in range(N+1)]
memo
```

```python
def knapsack(N, W, profit, weight):
    
    if N == 0 or W == 0:
        return 0
    
    if memo[N][W] >-1:
        return memo[N][W]
    
    if W < weight[N-1]:
        memo[N][W] =  knapsack(N - 1, W, profit, weight)
        return memo[N][W]
    
    memo[N][W] = max(profit[N-1] + knapsack(N - 1, W -weight[N-1] , profit, weight) , knapsack(N-1, W , profit, weight))
    
#     memo[N-1][W - weight[0]] = knapsack(N - 1, W -weight[n-1] , profit, weight) + profit[n-1]
#     memo[N-1][W] = knapsack(N-1, W , profit, weight)

    return memo[N][W]
```

```python
N = 3
W = 4
profit = [1, 2, 3]
weight = [4, 5, 1]
```

```python
def knapsack(n, W, val, wt):
    K = [[0 for x in range(W + 1)] for x in range(n + 1)]
 
    # Build table K[][] in bottom up manner
    for i in range(n + 1):
        for w in range(W + 1):
            if i == 0 or w == 0:
                K[i][w] = 0
            elif wt[i-1] <= w:
                K[i][w] = max(val[i-1]
                              + K[i-1][w-wt[i-1]],
                              K[i-1][w])
            else:
                K[i][w] = K[i-1][w]
    print(K)
    return K[n][W]
```

```python
knapsack(N, W, profit, weight)
```
