### DP

#### Recurrsion + Memo

```python
coins = [1,2,5,12,15]
amount = 11
n = len(coins)
```

```python
coins = [2]
amount = 3
n = len(coins)
```

```python

def coin_change_dp(coins, amount,n):
#     print(coins, n-1)
    
    n = len(coins)
    dp =[ [None for i in range(amount + 1)] for j in range(n+1)]

    def coin_change(coins, amount,n):
        # coins.sort()
        if n == 0 and amount > 0:
            dp[n][amount] =  float("inf")
            return dp[n][amount]


        if n >= 0 and amount == 0:
            dp[n][amount] =  0
            return dp[n][amount]

        if dp[n][amount]:
            return dp[n][amount]

        
        if coins[n-1] <= amount:

            dp[n][amount] = min(coin_change(coins, amount,n-1), coin_change(coins, amount - coins[n-1],n) + 1)

        else:
            dp[n][amount] = coin_change(coins, amount,n-1)

        return dp[n][amount]


    res = coin_change(coins, amount,n)
    if res != float("inf"):
        return res
    return -1
```

```python
coin_change_dp(coins, amount,n)
```

```python
def coin_change_tab(coins, amount,n):
    
    dp = [[None for x in range(amount + 1)] for y in range(n + 1)]

        
        
    for i in range(amount+1):
        dp[0][i] = float('inf')
        if i%coins[0]==0:
            dp[1][i] = i//coins[0]
        else:
            dp[1][i] = float('inf')
    for i in range(1, n+1):
        dp[i][0] = 0
        
        
    for i in range(2,n+1):
        for j in range(1,amount+1):
#             if n == 0 or amount == 0:
#                 dp[i][j] = 0
            if coins[i-1] <= j:
                dp[i][j] = min(dp[i-1][j],1 + dp[i][j-coins[i-1]] )
            else:
                dp[i][j] = dp[i-1][j]
                
    print(dp)

    if(dp[n][amount]==1000):
        return -1
    return dp[n][amount]
```
