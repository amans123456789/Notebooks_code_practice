```python
def min_eating_speed(piles, h):
    def can_finish(piles, k, h):
        time = 0
        for pile in piles:
            time += (pile + k - 1) // k  # Calculate the time needed to eat the pile
        return time <= h

    def binary_search(low, high, piles, h):
        if low >= high:
            return low

        mid = low + (high - low) // 2

        if can_finish(piles, mid, h):
            return binary_search(low, mid, piles, h)
        else:
            return binary_search(mid + 1, high, piles, h)

    low = 1
    high = max(piles)

    return binary_search(low, high, piles, h)
```
